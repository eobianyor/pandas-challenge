```python
# Dependencies and Setup
import pandas as pd
import os

# file paths
school_data_to_load = os.path.join("Resources", "schools_complete.csv")
student_data_to_load = os.path.join("Resources", "students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_and_student_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_and_student_data_complete.shape
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# DISTRICT SUMMARY

# Get total school count
Total_Schools = school_data['school_name'].count()
# Option B: Total_Schools = school_data['school_name'].nunique()

# Get total student count
Total_Students = student_data['student_name'].count()

# Get total budget
Total_Budget = school_data['budget'].sum()

# Average Math Score
Av_Math_Score = student_data['math_score'].mean()

# Average Reading Score
Av_Reading_Score = student_data['reading_score'].mean()

# Math pass percentage 70 plus
Math70Plus_df = student_data[student_data.math_score>=70]
MathPassPercentage = ((Math70Plus_df['student_name'].count())/(student_data['student_name'].count()) * 100)

# Reading pass percentage 70 plus
Reading70Plus_df = student_data[student_data.reading_score>=70]
ReadingPassPercentage = ((Reading70Plus_df['student_name'].count())/(student_data['student_name'].count()) * 100)

# Pass percentage for Math and Reading
Both70Plus_df = student_data[(student_data.math_score>=70) & (student_data.reading_score>=70)]
BothPassPercentage = ((Both70Plus_df['student_name'].count())/(student_data['student_name'].count()) * 100)

# Create a district Summary data frame
list_for_df = {"Total_Schools": [Total_Schools], "Total_Students": [Total_Students], "Total_Budget": [Total_Budget], "Av_Math_Score": [Av_Math_Score], 
               "Av_Reading_Score": [Av_Reading_Score], "MathPass%": [MathPassPercentage], "ReadingPass%": [ReadingPassPercentage], "CombinedPass%": [BothPassPercentage]}

District_Summary_df = pd.DataFrame(list_for_df)
District_Summary_df.head()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# SCHOOL SUMMARY

# Sort school data and create a dataset grouped by schools
sorted_school_data = school_data.sort_values('school_name')
school_group_grpby = school_and_student_data_complete.groupby('school_name')

# Get school name and type, no. of student budget and budget per student
school_name = (sorted_school_data['school_name'].unique())
school_type = (sorted_school_data['type'].tolist())
Total_Students = (sorted_school_data['size'].unique())
Total_School_Budget = (sorted_school_data['budget'].unique())
Per_Student_Budget = (((sorted_school_data['budget'])/(sorted_school_data['size'])).unique())

# Get mean reading and math scores
Av_School_math_score = school_group_grpby['math_score'].mean()
Av_School_read_score = school_group_grpby['reading_score'].mean()
school_group_grpby.count().head(15)

# students passing math broken down by school
All_schs_Math70plus = school_and_student_data_complete[school_and_student_data_complete.math_score>=70]
sch_grp_Math70plus_grpby = All_schs_Math70plus.groupby('school_name')
sch_grp_Math70plus_grpby.count().head(15)
sch_grp_Math70plus_grpby['math_score'].count()
school_group_grpby['math_score'].count()
MathPassPercentage_per_Sch = ((sch_grp_Math70plus_grpby['math_score'].count())/(school_group_grpby['math_score'].count()) * 100)
MathPassPercentage_per_Sch.shape

# students passing reading broken down by school
All_schs_Read70plus = school_and_student_data_complete[school_and_student_data_complete.reading_score>=70]
sch_grp_Read70plus_grpby = All_schs_Read70plus.groupby('school_name')
sch_grp_Read70plus_grpby.count().head(15)
sch_grp_Read70plus_grpby['reading_score'].count()
school_group_grpby['reading_score'].count()
ReadPassPercentage_per_Sch = ((sch_grp_Read70plus_grpby['reading_score'].count())/(school_group_grpby['reading_score'].count()) * 100)
ReadPassPercentage_per_Sch.shape

# students passing both math and reading broken down by school
All_schs_Both70plus = school_and_student_data_complete[(school_and_student_data_complete.math_score>=70) & (school_and_student_data_complete.reading_score>=70)]
sch_grp_Both70plus_grpby = All_schs_Both70plus.groupby('school_name')
sch_grp_Both70plus_grpby.count().head(15)
sch_grp_Both70plus_grpby['reading_score'].count()
school_group_grpby['reading_score'].count()
BothPassPercentage_per_Sch = ((sch_grp_Both70plus_grpby['reading_score'].count())/(school_group_grpby['reading_score'].count()) * 100)
BothPassPercentage_per_Sch.head(15)

# Creating a new data frame with School Summary
list2_for_df = {"School_type": (school_type), "School_name": (school_name), "Total_Students": (Total_Students), "School_Budget": (Total_School_Budget), 
                "Per_Student_Budget": (Per_Student_Budget), "Average_Math_Score": (Av_School_math_score), "Average_Reading_Score": (Av_School_read_score), 
                "Math_Pass%": (MathPassPercentage_per_Sch), "Reading_Pass%": (ReadPassPercentage_per_Sch), "Overall_Pass%": (BothPassPercentage_per_Sch)}

School_Summary_df = pd.DataFrame(list2_for_df)
School_Summary_df.head(15)
```
